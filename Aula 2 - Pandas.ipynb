## Exercícios

1. Considere a existência de três tabelas distintas:
* customer.csv : Possui a informação dos clientes em duas colunas: customer id  customer name
* products.csv : Contém informação dos produtos vendidos pela empresa em três colunas - p_id (product id), product (name) e price
* sales.csv : Contém informações das vendas realizadas em seis colunas - sale_id, c_id (customer id), p_id (product_id), qty (quantity sold), store (name)

```python
customer = pd.read_csv('./data/customer.csv')
products = pd.read_csv('./data/products.csv')
sales = pd.read_csv('./data/sales.csv')
```

```python
df_vendas = sales.merge(products, 
           how="left",
          # on="p_id",
           suffixes=('_sales', '_product'))

df_vendas = df_vendas.merge(customer, 
           how="left",
           on="c_id",
           suffixes=('_sales', '_costumer'))

df_vendas
```

Conhecendo as bases responda:

a) Quais foram os produtos vendidos? 


```python
df_vendas['product'].unique()
```

b) E os não vendidos?

```python
products[~products.p_id.isin(df_vendas.p_id)][['product']]
```

c) Quais foram as cinco maiores vendas? Salve essas vendas em um arquivo excel.

```python
df_vendas['valor_venda'] = df_vendas['qty'] * df_vendas['price']
df_vendas.sort_values(by='valor_venda', ascending=False).head()
```

d) Liste a quantidade vendida de cada produto. Utilize um loop for para isso.

```python
for produto in df_vendas['product'].unique():
    print(f"{produto} teve {df_vendas[df_vendas['product'] == produto]['qty'].sum()} quantidade(s) vendida(s)")
```

e) Liste a quantidade vendida de cada loja.

```python
for loja in df_vendas['store'].unique():
    print(f"{loja} - teve {df_vendas[(df_vendas['store'] == loja)]['qty'].sum()} quantidade(s) vendida(s)")
```

f) Liste a quantidade vendida de cada produto por loja.

```python
for loja in df_vendas['store'].unique():
    for produto in df_vendas['product'].unique():
        print(f"{loja} - {produto} teve {df_vendas[(df_vendas['product'] == produto) & (df_vendas['store'] == loja)]['qty'].sum()} quantidade(s) vendida(s)")
```
