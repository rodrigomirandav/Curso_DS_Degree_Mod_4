b. Sabendo que as cinco colunas correspondem, nessa ordem, a: 
   * 1. sepal length (cm)
   * 2. sepal width (cm)
   * 3. petal length (cm)
   * 4. petal width (cm)
   * 5. class: <br>
        - Iris Setosa <br>
        - Iris Versicolor <br>
        - Iris Virginica <br>

Leia novamente o arquivo passando o nome das colunas como argumento.

```python
df_iris = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data',
                      sep=',',
                      header=None,
                      names=['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'class'])
df_iris
```

c. Crie uma coluna para o volume sabendo que ele é calculado por (pi x petallength x sepal_length^2)/3

```python
import math
df_iris["volume"] = (math.pi * df_iris["petal_length"] * (df_iris["sepal_length"]**2))/3
df_iris
```

d. Calcule a média de cada atributo das flores.

```python
df_iris.loc[:,['sepal_length','sepal_width','petal_length','petal_width','volume']].mean()
```

e. Novamente, calcule a média de cada atributo das flores só que agora faça-o separadamente para cada um dos tipos de íris indicados na coluna "class". 
Utilize o loop for para reduzir quantidade de linhas. <br>
Existe diferença entre as médias de cada um dos atributos das flores?

```python
for tipo_iris in ['Iris-setosa', 'Iris-versicolor', 'Iris-virginica']:
    print(tipo_iris)
    print(df_iris[df_iris['class'] == tipo_iris].loc[:,['sepal_length','sepal_width','petal_length','petal_width','volume']].mean())
    print()
    
    
# Sim existe diferença nas médias dos atributos por flores
```

f. Adicione uma nova coluna com o valor da média de 'sepal length' apenas para a íris da classe Iris-setosa.

```python
media_sepal_length = df_iris[df_iris['class'] == 'Iris-setosa']['sepal_length'].mean()

df_iris.loc[df_iris['class'] == 'Iris-setosa','media_sepal_length'] = media_sepal_length
df_iris
```

g. Adicione uma coluna com a média de 'sepal length' de tal forma que cada tipo de íris tenha seu respectivo valor de média. Utilize o loop for e faça isso sem utilizar métodos que não foram ensinados na aula de hoje.

```python
for tipo in ['Iris-setosa', 'Iris-versicolor', 'Iris-virginica']:
    media_sepal_length = df_iris[df_iris['class'] == tipo]['sepal_length'].mean()

    df_iris.loc[df_iris['class'] == tipo,'media_sepal_length'] = media_sepal_length

df_iris
```
